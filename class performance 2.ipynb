```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
ds = pd.read_csv('student_performance.csv')
```

```python
ds
```

```python
ds = ds.drop(['Name', 'StudentID'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender = LabelEncoder()
le_parentalsupport = LabelEncoder()
```

```python
ds['Gender']=le_gender.fit_transform(ds['Gender'])
ds['ParentalSupport'] = le_parentalsupport.fit_transform(ds['ParentalSupport'])
```

```python
ds
```

```python
x = ds.drop('FinalGrade', axis = 1)
y = ds['FinalGrade']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
